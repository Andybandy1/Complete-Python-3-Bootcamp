```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('/content/AAL.csv')
```

```python
dataset.describe()
```

```python
x = dataset[['High','Low','Open']].values
y = dataset['Close'].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
regressor =  LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
print(regressor.coef_)
```

```python
print(regressor.intercept_)
```

```python
predicted = regressor.predict(x_test)
```

```python
dframe = pd.DataFrame({'Actual':y_test.flatten(), 'Predicted':predicted.flatten()})
```

```python
dframe.head(10)
```

```python
dframe.tail(10)
```

```python
import math
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predicted))
print('Mean Squared Root:', metrics.mean_squared_error(y_test, predicted))
print('Root Mean Squared Error:', math.sqrt(metrics.mean_squared_error(y_test, predicted)))
```
